## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
############# Working
# basic calculations to obtain number of unique items, average price, etc.


# Number of Unique Items
unique_items = purchase_data["Item ID"].nunique()
# print(unique_items)

# Average Price
average_price = purchase_data["Price"].mean()
# print(average_price)

# Number of purchases
num_pur = purchase_data["Purchase ID"].count()
# print(num_pur)

# Total Revenue
total_rev = purchase_data["Price"].sum()
# print(total_rev)

# Puts all the pieces together to display (data frame)
Purchasing_Analysis_total = pd.DataFrame({"Number of Unique Items": [unique_items], 
                                          "Average Price": [average_price], 
                                          "Number of purchases": [num_pur], 
                                          "Total Revenue": [total_rev]})
# Purchasing_Analysis_total

#Formating
Purchasing_Analysis_total["Total Revenue"] = Purchasing_Analysis_total["Total Revenue"].map("${:,.2f}".format)
Purchasing_Analysis_total["Average Price"] = Purchasing_Analysis_total["Average Price"].map("${:.2f}".format)

# Screen output
Purchasing_Analysis_total
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
############# Working

# grabs gender, price and screen name
Purchasing_Analysis_gender = purchase_data[["Gender", "Price", "SN"]]
# Purchasing_Analysis_gender.head()


# creates group by screen name
gender_demographics5 = Purchasing_Analysis_gender.groupby(["SN"])
# age_demographics.head(780)
gender_demographics6 = gender_demographics5["Gender"].unique()
# print(gender_demographics1)
gender_demographics6_df = pd.DataFrame({"Gender": gender_demographics6})
# gender_demographics_df.head()
gender_demographics_df_countxx = gender_demographics6_df["Gender"].value_counts()
# print(gender_demographics_df_countxx)

# group Purchasing_Analysis_gender by gender
Purchasing_Analysis_gender_group = Purchasing_Analysis_gender.groupby(["Gender"])

# The average price
average_price_g = Purchasing_Analysis_gender_group["Price"].mean()
# print(average_price_g)

# the total price
total_price_g = Purchasing_Analysis_gender_group["Price"].sum()
# print(total_price_g)

# The Purchase count
purchase_price_g = Purchasing_Analysis_gender["Gender"].value_counts()
# print(purchase_price_g)

 # create the datafram so that the data can be put together  
Purchasing_Analysis_gender_df = pd.DataFrame({"Gender":["Female", "Male", "Other / Non-Disclosed"],
                                              "temp_index":["1", "2", "3"],
                                              "Purchase Count": purchase_price_g, 
                                             "Average Purchase Price": average_price_g,
                                             "Total Purchase Value": total_price_g})

# % of players
gender_df_sum2 = gender_df.set_index("Percentage of Players")
gender_df_sum2.index.name = None
# gender_df_sum2    

# temp index (1,2,3) to allow data to not be set on gender
gender_df_sum3 = Purchasing_Analysis_gender_df.set_index("temp_index")
gender_df_sum3.index.name = None
# gender_df_sum3   

# putting the % of players and the Purchasing_Analysis_gender_df into one table 
     #using Gender to match it up
merge_table_pa = pd.merge(gender_df_sum3, gender_df_sum2, on="Gender")
# merge_table_pa.head()

pa_percentage = round((merge_table_pa["Total Purchase Value"]/merge_table_pa["Total Count"]),2)
# pa_percentage
merge_table_pa.loc[:,"Avg Total Purchase per Person"] = pa_percentage
# merge_table_pa.head()


merge_table_pa2 = merge_table_pa.set_index("Gender")
# merge_table_pa2.head()
del merge_table_pa2['Total Count']
# merge_table_pa2

# Formats the data
merge_table_pa2["Average Purchase Price"] = merge_table_pa2["Average Purchase Price"].map("${:.2f}".format)
merge_table_pa2["Total Purchase Value"] = merge_table_pa2["Total Purchase Value"].map("${:,.2f}".format)
merge_table_pa2["Avg Total Purchase per Person"] = merge_table_pa2["Avg Total Purchase per Person"].map("${:.2f}".format)

# Screen output
merge_table_pa2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
############# Working

# grabs screen name and age
age_demo = purchase_data[["SN", "Age"]]

# creates group by screen name
age_demographics = age_demo.groupby(["SN"])
# age_demographics.head(780)

age_demographics1 = age_demographics["Age"].unique()
# print(age_demographics1)

age_demographics_df = pd.DataFrame({"Age": age_demographics1})
# age_demographics_df

Total_age_demographics_df = age_demographics_df["Age"].count()
# print(Total_age_demographics_df)


# create the bins for age (last bin just needs to be larger then the oldest person)
# bins = [0, 10, 15, 20, 25, 30, 35, 40, 9999]
bins = [0, 9, 14, 19, 24, 29, 34, 39, 9999]

# Create the names for the 8 bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_demographics_df["Age Group"] = pd.cut(age_demographics_df["Age"], bins, labels=group_names)
age_demographics_df.head(50)

# grabs the age and age group from the age_demographics_df dataframe
age_demo2 = age_demographics_df[["Age", "Age Group"]]
age_demographics2 = age_demo2.groupby(["Age Group"])
age_demographics2 = age_demographics2["Age"].count()
# age_demographics2


age_demographics2_df = pd.DataFrame({"Age": age_demographics2})
# age_demographics2_df

age_demographics2_renamed_df = age_demographics2_df.rename(columns={"Age":"Total Count"})
# age_demographics2_renamed_df

# create the Percentage of Players
age_percentage = round((age_demographics2_renamed_df["Total Count"]/Total_age_demographics_df)*100,2)
# adds the Percentage of Players into the age_demographics2_renamed_df dataframe
age_demographics2_renamed_df["Percentage of Players"] = age_percentage

# Screen output
age_demographics2_renamed_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#### WORKING

# group item id and item name
Purchasing_Analysis_Most_Popular_Items_group = purchase_data.groupby(["Item ID", "Item Name"])
#count the item id's
psmpig_count = Purchasing_Analysis_Most_Popular_Items_group["Item ID"].count()
# total (sum) the price
psmpig_sum = Purchasing_Analysis_Most_Popular_Items_group["Price"].sum()
#devide the total price by the number of item id's
psmpig_orig_price = psmpig_sum / psmpig_count

# put results into a dataframe
most_popular_items_df = pd.DataFrame({"Purchase Count": psmpig_count, 
                                     "Item Price": psmpig_orig_price,
                                     "Total Purchase Value": psmpig_sum})
 
# most_popular_items_df
# Sort on Purchase Count
most_popular_items_df_sort = most_popular_items_df.sort_values("Purchase Count", ascending=False)
# most_popular_items_df_sort.head()


#Formating
most_popular_items_df_sort["Item Price"] = most_popular_items_df_sort["Item Price"].map("{:.2f}".format)
most_popular_items_df_sort["Total Purchase Value"] = most_popular_items_df_sort["Total Purchase Value"].map("${:.2f}".format)

# Screen output (first 5 records)
most_popular_items_df_sort.head()
```
